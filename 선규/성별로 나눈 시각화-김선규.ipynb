```python
import matplotlib.pyplot as plt

def visualize(df, month, category, name, gender):
    month_title = month//100
    # 해당 조건에 맞는 데이터 필터링
    filtered_data = df[(df['조사연월'] == month) & (df['가구주_성별코드'] == gender)][category].sum()

    if filtered_data.empty:
        print(f"{month}월에 해당하는 {gender} 데이터가 없습니다.")
        return

    # 막대 그래프(bar plot) 그리기
    plt.figure(figsize=(12, 8))
    bars = plt.bar(filtered_data.index, filtered_data.values, color='blue' if gender == '남성' else 'pink')
    for bar in bars:
        plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 100, int(bar.get_height()), ha='center', fontsize=10)
    plt.title(f"{month_title}년도 {gender}의 {name}", fontsize=20)
    plt.xlabel(f"{name}", fontsize=20)
    plt.ylabel(f"{name}의 비용", fontsize=20)
    plt.ylim(80000)
    plt.xticks(rotation=90)
    
    plt.tight_layout()
    plt.show()
```

```python
visualize(cleaned_single_expend, 202100, cmnct, '통신', '남성')
```
